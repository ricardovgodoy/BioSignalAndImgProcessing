# **Code Explanation for Types of Discrete Signals**
## **1. Generating the Time Axis** *n*

*   We create an array of integer values from -10 to 10 using `np.arange(-10, 10, 1)`.
*   Since discrete signals only exist at integer values, this vector defines the instants where the samples are evaluated.

## **2. Creating the Discrete Signals**
### **Unit Impulse ùõø[ùëõ]**
*   It uses `np.where(n == 0, 1, 0)`, which sets 1 at ùëõ=0 and 0 at all other points.
### **Unit Step ùë¢[ùëõ]**
*   `np.where(n >= 0, 1, 0)` sets 1 for n‚â•0 and 0 for ùëõ<0.
### **Discrete Sinusoid ùë•[ùëõ]=sin‚Å°(0.2ùúãùëõ)**
*   We display a sinusoid using `np.sin(0.2 * np.pi * n)`.

## **3. Visualization with matplotlib**

*   We use `plt.stem()` for discrete plots (it uses markers instead of continuous lines).

*   Each subplot shows a different signal with title, axes, and grid.

# **Expected Output**
This code will generate three plots:
*   Unit Impulse: A single point at n=0.
*   Unit Step: A sequence with value 0 before ùëõ=0 and 1 afterward.
*   Discrete Sinusoid: A periodic sequence representing a sampled sinusoid.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definition of the discrete-time axis
n = np.arange(-10, 10, 1)  # n interval from -10 to 10

# 1. Unit impulse Œ¥[n]
impulso = np.where(n == 0, 1, 0)

# 2. Unit step u[n]
degrau = np.where(n >= 0, 1, 0)

# 3. Discrete sinusoidal signal x[n] = sin(0.2œÄn)
senoide = np.sin(0.2 * np.pi * n)

# Create figure for the three signals
plt.figure(figsize=(10, 6))

# Function to remove the baseline
def plot_stem(n, x, title):
    markerline, stemlines, baseline = plt.stem(n, x)
    plt.setp(stemlines, 'color', 'black')  # Black vertical lines
    plt.setp(markerline, 'color', 'blue')  # Blue points
    baseline.set_visible(False)  # Hide the red baseline
    plt.title(title)
    plt.xlabel("n")
    plt.ylabel("Amplitude")
    plt.grid()

# Unit Impulse plot
plt.subplot(3, 1, 1)
plot_stem(n, impulso, "Unit Impulse Œ¥[n]")

# Unit Step plot
plt.subplot(3, 1, 2)
plot_stem(n, degrau, "Unit Step u[n]")

# Discrete Sinusoid plot
plt.subplot(3, 1, 3)
plot_stem(n, senoide, "Sinusoidal Signal x[n] = sin(0.2œÄn)")

# Adjust layout and show
plt.tight_layout()
plt.show()
```

## **Code Explanation for an Ideal Delay and Advance System**
1.   We create a discrete signal `ùë•[ùëõ]=sin‚Å°(0.2ùúãùëõ)`.
2.   Time delay `(ùë•[ùëõ‚àí3])`: The signal is shifted to the right.
3.   Time advance `(ùë•[ùëõ+3])`: The signal is shifted to the left.
4.   Visualization:
*   The original signal is blue.
*   The delayed signal is green.
*   The advanced signal is red.

## **What will be shown in the plots**
1.   The delayed signal `(ùë•[ùëõ‚àí3])` starts later (shifted to the right).
2.   The advanced signal `(ùë•[ùëõ+3])` happens earlier (shifted to the left).


```python
import numpy as np
import matplotlib.pyplot as plt

# Define discrete time
n = np.arange(-10, 11, 1)  # Discrete-time interval

# Create an original signal: discrete sinusoid
x_n = np.sin(0.2 * np.pi * n)

# Apply delay and advance
n_atraso = 3  # Number of samples for delay
n_avanco = 3  # Number of samples for advance

x_atrasado = np.sin(0.2 * np.pi * (n - n_atraso))  # x[n - 3] (delay)
x_avancado = np.sin(0.2 * np.pi * (n + n_avanco))  # x[n + 3] (advance)

# Create figure for plots
plt.figure(figsize=(10, 6))

# Original Signal plot
plt.subplot(3, 1, 1)
plt.stem(n, x_n, linefmt='b-', markerfmt='bo', basefmt="k-")
plt.title("Original Signal x[n] = sin(0.2œÄn)")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid()

# Delayed Signal plot
plt.subplot(3, 1, 2)
plt.stem(n, x_atrasado, linefmt='g-', markerfmt='go', basefmt="k-")
plt.title(f"Delayed Signal x[n-{n_atraso}]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid()

# Advanced Signal plot
plt.subplot(3, 1, 3)
plt.stem(n, x_avancado, linefmt='r-', markerfmt='ro', basefmt="k-")
plt.title(f"Advanced Signal x[n+{n_avanco}]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid()

# Adjust layout and show
plt.tight_layout()
plt.show()
```

## **Code Explanation to Illustrate Periodic Sampling**
1.   **Continuous Signal Generation**:
*   The continuous signal is defined as `ùë•(ùë°)=sin‚Å°(2ùúã√ó5ùë°)`, that is, a sinusoid with 5 Hz frequency.
*   A time vector `t_cont` with 1000 points is used to represent the signal continuously.

2.   **Sampling with fs = 20 Hz**:
*   Samples are collected every 1/20 second.
*   The resulting points are shown with red markers.

3.   **Sampling with fs = 40 Hz**:
*   Now we sample at twice the rate, collecting points every 1/40 second.
*   The resulting points (in green) highlight a more accurate representation of the continuous signal.

## **Conclusion**
Increasing the sampling rate improves how well the discrete signal approximates the continuous signal.


```python
import numpy as np
import matplotlib.pyplot as plt

# Continuous signal parameters
f_signal = 7         # Signal frequency (Hz)
T = 1                # Signal duration (1 second)
t_cont = np.linspace(0, T, 1000, endpoint=False)  # High-resolution time vector
x_cont = np.sin(2 * np.pi * f_signal * t_cont)     # Continuous signal: 5 Hz sine wave

# Defining two sampling rates
fs1 = 20             # First sampling rate (20 Hz)
fs2 = 40             # Second sampling rate (40 Hz, double 20 Hz)

# Sampling the signal
t_sample1 = np.arange(0, T, 1/fs1)
x_sample1 = np.sin(2 * np.pi * f_signal * t_sample1)

t_sample2 = np.arange(0, T, 1/fs2)
x_sample2 = np.sin(2 * np.pi * f_signal * t_sample2)

# Plotting the results
plt.figure(figsize=(10, 6))

# Subplot with fs = 20 Hz
plt.subplot(2, 1, 1)
plt.plot(t_cont, x_cont, label='Continuous Signal', color='blue')
plt.stem(t_sample1, x_sample1, linefmt='r-', markerfmt='ro', basefmt='k-', label='Samples (fs = 20 Hz)')
plt.title('Sampling with fs = 20 Hz')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Subplot with fs = 40 Hz
plt.subplot(2, 1, 2)
plt.plot(t_cont, x_cont, label='Continuous Signal', color='blue')
plt.stem(t_sample2, x_sample2, linefmt='g-', markerfmt='go', basefmt='k-', label='Samples (fs = 40 Hz)')
plt.title('Sampling with fs = 40 Hz (Double Rate)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# **Code for an impulse train**
An impulse train is a sequence of impulses regularly spaced in time.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definition of discrete time
n = np.arange(-20, 21)  # Discrete-time vector
N = 5  # Impulse train period

# Creating the discrete impulse train (impulses every N samples)
impulsos = np.zeros(len(n))
impulsos[n % N == 0] = 1  # Sets impulses spaced by N samples

# Creating the plot
plt.figure(figsize=(8, 4))
markerline, stemlines, baseline = plt.stem(n, impulsos)

# Removing the red connecting line
plt.setp(stemlines, 'color', 'black')  # Set vertical lines to black
plt.setp(markerline, 'color', 'blue')  # Set markers to blue
baseline.set_visible(False)  # Hide the red horizontal line

# Visual adjustments
plt.title("Discrete Impulse Train")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

# **Aliasing example**


```python
import numpy as np
import matplotlib.pyplot as plt

# Continuous signal parameters
f_signal = 5        # Original signal frequency (Hz)
T = 1               # Total signal duration (1 second)
t_cont = np.linspace(0, T, 1000, endpoint=False)  # High-resolution continuous-time vector
x_cont = np.sin(2 * np.pi * f_signal * t_cont)      # Continuous signal: 5 Hz sine wave

# Define two sampling rates
fs_high = 50        # High sampling rate (50 Hz) - satisfies the Nyquist criterion
fs_low = 8          # Low sampling rate (8 Hz) - insufficient for 5 Hz (aliasing)

# Signal sampling with fs = 50 Hz (high quality)
t_sample_high = np.arange(0, T, 1/fs_high)
x_sample_high = np.sin(2 * np.pi * f_signal * t_sample_high)

# Signal sampling with fs = 8 Hz (aliasing may occur)
t_sample_low = np.arange(0, T, 1/fs_low)
x_sample_low = np.sin(2 * np.pi * f_signal * t_sample_low)

# Theoretical alias calculation:
# For f_signal = 5 Hz and fs = 8 Hz, the aliased frequency is:
#    f_alias = |5 - 8| = 3 Hz
# Therefore, the ideal reconstruction (with aliasing) is:
f_alias = 3  # Aliased frequency
x_alias = -np.sin(2 * np.pi * f_alias * t_cont)

# Plotting the results
plt.figure(figsize=(12, 10))

# Subplot 1: Continuous signal with fs = 50 Hz sampling (high rate)
plt.subplot(3, 1, 1)
plt.plot(t_cont, x_cont, label='Continuous Signal', color='blue')
plt.stem(t_sample_high, x_sample_high, linefmt='r-', markerfmt='ro', basefmt='k-', label='Samples (fs = 50 Hz)')
plt.title('Sampling with fs = 50 Hz')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Subplot 2: Continuous signal with fs = 8 Hz sampling (low fs, with aliasing)
plt.subplot(3, 1, 2)
plt.plot(t_cont, x_cont, label='Continuous Signal', color='blue')
plt.stem(t_sample_low, x_sample_low, linefmt='r-', markerfmt='ro', basefmt='k-', label='Samples (fs = 8 Hz)')
plt.title('Sampling with fs = 8 Hz (Aliasing Occurs)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Subplot 3: Theoretical reconstruction from points sampled at fs = 8 Hz
plt.subplot(3, 1, 3)
plt.plot(t_cont, x_alias, label='Reconstructed Signal (Aliasing)', color='magenta', linewidth=2)
plt.stem(t_sample_low, x_sample_low, linefmt='r-', markerfmt='ro', basefmt='k-', label='Samples (fs = 8 Hz)')
plt.title('Signal Reconstruction with Aliasing')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# **ECG**

### **Generating a synthetic ECG signal**:
The `generate_ecg` function creates a signal composed of Gaussian peaks (simulating QRS complexes) positioned at the instants defined in `beat_times`.

A 60 Hz sinusoidal interference is added to the signal. This interference, when sampled at 5 Hz, will undergo aliasing (the 60 Hz signal will appear with a distorted frequency).

### **Comparison between low and high sampling:**
Two signals are generated:

*   ECG with `fs_low` (5 Hz): It may show aliasing, because 5 Hz is not twice the maximum frequency (the 60 Hz interference should require at least 120 Hz to avoid aliasing).
*   ECG with `fs_high` (300 Hz): Proper sampling that preserves all signal details.


### **RR interval detection:**
The `find_peaks` function is used to detect ECG peaks (QRS complexes).

From the peak times, RR intervals are calculated.

A simple criterion is applied to detect arrhythmias: if an interval deviates more than 20% from the median interval, it indicates an abnormal beat.

### **Visualization:**
The signals with different sampling rates and the signal with detected peaks are plotted, making it easier to visualize aliasing effects and identify abnormal intervals.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# ---------------------------
# Simulation parameters
# ---------------------------
duration = 10.0          # signal duration in seconds
fs_high = 300           # high sampling rate (Hz) - no aliasing
fs_low = 5              # low sampling rate (Hz) - may cause aliasing
t_high = np.linspace(0, duration, int(duration * fs_high), endpoint=False)
t_low  = np.linspace(0, duration, int(duration * fs_low), endpoint=False)

# Defining beat times (in seconds)
# Note: the interval between 4.5s and 6.0s (1.5s) simulates an arrhythmia (abnormal pause)
beat_times = [0.5, 1.5, 2.5, 3.5, 4.5, 6.0, 7.0, 8.0, 9.0]

# ---------------------------
# Function to generate synthetic ECG signal
# ---------------------------
def generate_ecg(t, beat_times, amplitude=1.0, sigma=0.03, interference_amp=0.2, interference_freq=60):
    """
    Generates a synthetic ECG signal composed of Gaussian peaks representing QRS complexes,
    with high-frequency sinusoidal interference (e.g., 60 Hz).

    Parameters:
      - t: time vector
      - beat_times: list of heartbeat times
      - amplitude: amplitude of QRS peaks
      - sigma: Gaussian standard deviation (peak width)
      - interference_amp: interference amplitude
      - interference_freq: interference frequency (Hz)

    Returns:
      - ECG signal
    """
    ecg = np.zeros_like(t)
    for bt in beat_times:
        ecg += amplitude * np.exp(-((t - bt)**2) / (2 * sigma**2))
    # Add high-frequency interference
    interference = interference_amp * np.sin(2 * np.pi * interference_freq * t)
    return ecg + interference

# Generating ECG signals with different sampling rates
ecg_high = generate_ecg(t_high, beat_times)
ecg_low  = generate_ecg(t_low, beat_times)

# ---------------------------
# Plotting ECG signals
# ---------------------------
plt.figure(figsize=(12, 8))

# Signal with low sampling rate (possible aliasing)
plt.subplot(2, 1, 1)
plt.plot(t_low, ecg_low, 'b.-')
plt.title(f'ECG with low sampling ({fs_low} Hz) - With aliasing')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Signal with high sampling rate (proper capture)
plt.subplot(2, 1, 2)
plt.plot(t_high, ecg_high, 'r-')
plt.title(f'ECG with high sampling ({fs_high} Hz) - Without aliasing')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

# ---------------------------
# Arrhythmia detection: extracting RR intervals
# ---------------------------
# We will use the high-sampling-rate signal for more accurate detection.
# We use "find_peaks" with a minimum height to identify peaks (QRS)
peaks, properties = find_peaks(ecg_high, height=0.5, distance=fs_high*0.4)
peak_times = t_high[peaks]
RR_intervals = np.diff(peak_times)  # intervals in seconds between beats

print("RR intervals (in seconds):")
print(RR_intervals)

# Simple criterion to detect arrhythmia: if interval varies more than 20% from the median
median_RR = np.median(RR_intervals)
threshold = 0.2 * median_RR
arrhythmia_indices = np.where(np.abs(RR_intervals - median_RR) > threshold)[0]

print("
Indices of intervals with arrhythmia (abnormal):")
print(arrhythmia_indices)
print("Abnormal RR intervals:")
print(RR_intervals[arrhythmia_indices])

# Plotting ECG signal (high sampling) with detected peaks
plt.figure(figsize=(10, 4))
plt.plot(t_high, ecg_high, label='ECG')
plt.plot(peak_times, ecg_high[peaks], "x", label="Detected peaks", markersize=10)
plt.title('ECG Peak Detection (QRS)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

# **Code Explanation:**
### **ECG Signal Generation:**

The ECG signal is simulated as a combination of sinusoids with frequencies of 0.5 Hz, 5 Hz, and 40 Hz.

### **Sampling:**

The signal is sampled at three different rates: 100 Hz, 200 Hz, and 50 Hz.

### **Aliasing Analysis:**

Separate plots are generated for the original signal and each sampled signal, allowing visualization of the aliasing effect at each sampling rate.

### **Anomaly Detection:**

Peaks (heartbeats) are detected in the original signal and in each sampled signal.


## **Results:**
### **Aliasing Analysis:**

The original signal will be plotted at the top.

Below it, three plots will show signals sampled at 100 Hz, 200 Hz, and 50 Hz.

The signal sampled at 50 Hz will show distortion due to aliasing.

### **Anomaly Detection**:

The original signal will show correctly detected peaks.

Signals sampled at 100 Hz and 200 Hz will show peak detection similar to the original.

The signal sampled at 50 Hz may present peak-detection errors due to aliasing, with a different peak count.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# 1. Simulation of an ECG signal
def generate_ecg_signal(duration, sampling_rate):
    t = np.linspace(0, duration, int(duration * sampling_rate), endpoint=False)
    # Simulation of an ECG signal with variable frequency components
    ecg_signal = (1.0 * np.sin(2 * np.pi * 0.5 * t) +  # Low-frequency component
                  0.5 * np.sin(2 * np.pi * 5 * t) +    # Mid-frequency component
                  0.2 * np.sin(2 * np.pi * 40 * t))     # High-frequency component
    return t, ecg_signal

# Signal parameters
duration = 1.0  # Signal duration in seconds
sampling_rate_original = 1000  # Original sampling rate (Hz)
t, ecg_signal = generate_ecg_signal(duration, sampling_rate_original)

# 2. Signal sampling
def sample_signal(signal, original_rate, new_rate):
    step = int(original_rate / new_rate)
    sampled_signal = signal[::step]
    sampled_time = t[::step]
    return sampled_time, sampled_signal

# Sampling rates for testing
sampling_rates = [100, 200, 50]  # Hz (50 Hz is below the Nyquist frequency for 40 Hz)
sampled_signals = []

for rate in sampling_rates:
    sampled_time, sampled_signal = sample_signal(ecg_signal, sampling_rate_original, rate)
    sampled_signals.append((sampled_time, sampled_signal, rate))

# 3. Aliasing analysis (separate plots for each frequency)
def plot_aliasing_effect_separate(original_signal, sampled_signals):
    plt.figure(figsize=(12, 8))

    # Original signal plot
    plt.subplot(len(sampled_signals) + 1, 1, 1)
    plt.plot(t, original_signal, label='Original Signal (1000 Hz)', color='black')
    plt.title('Original ECG Signal')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.legend()

    # Plots of sampled signals
    for i, (time, signal, rate) in enumerate(sampled_signals):
        plt.subplot(len(sampled_signals) + 1, 1, i + 2)
        plt.plot(time, signal, label=f'Signal Sampled at {rate} Hz', color='blue')
        plt.title(f'Signal Sampled at {rate} Hz')
        plt.xlabel('Time [s]')
        plt.ylabel('Amplitude')
        plt.legend()

    plt.tight_layout()
    plt.show()

plot_aliasing_effect_separate(ecg_signal, sampled_signals)

# 4. Anomaly detection (separate plots for each frequency)
def detect_irregular_beats(signal, time, threshold=0.5):
    peaks, _ = find_peaks(signal, height=threshold)
    peak_times = time[peaks]
    return peak_times, len(peaks)  # Returns peak times and peak count

def plot_anomaly_detection_separate(original_signal, sampled_signals):
    plt.figure(figsize=(12, 8))

    # Original signal plot with detected peaks
    plt.subplot(len(sampled_signals) + 1, 1, 1)
    plt.plot(t, original_signal, label='Original Signal (1000 Hz)', color='black')
    original_peaks, original_peak_count = detect_irregular_beats(original_signal, t)
    plt.plot(original_peaks, original_signal[np.array(original_peaks * sampling_rate_original, dtype=int)], 'x', label=f'Detected Peaks: {original_peak_count}', color='red')
    plt.title('Original ECG Signal with Detected Peaks')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.legend()

    # Plots of sampled signals with detected peaks
    for i, (time, signal, rate) in enumerate(sampled_signals):
        plt.subplot(len(sampled_signals) + 1, 1, i + 2)
        plt.plot(time, signal, label=f'Signal Sampled at {rate} Hz', color='blue')
        peaks, peak_count = detect_irregular_beats(signal, time)
        plt.plot(peaks, signal[np.array(peaks * rate, dtype=int)], 'o', label=f'Detected Peaks: {peak_count}', color='red')
        plt.title(f'Signal Sampled at {rate} Hz with Detected Peaks')
        plt.xlabel('Time [s]')
        plt.ylabel('Amplitude')
        plt.legend()

    plt.tight_layout()
    plt.show()

plot_anomaly_detection_separate(ecg_signal, sampled_signals)

# 5. Peak count for each frequency
def print_peak_counts(original_signal, sampled_signals):
    # Peak count in original signal
    _, original_peak_count = detect_irregular_beats(original_signal, t)
    print(f"Peaks detected in original signal (1000 Hz): {original_peak_count}")

    # Peak count in sampled signals
    for time, signal, rate in sampled_signals:
        _, peak_count = detect_irregular_beats(signal, time)
        print(f"Peaks detected in signal sampled at {rate} Hz: {peak_count}")

print_peak_counts(ecg_signal, sampled_signals)
```
