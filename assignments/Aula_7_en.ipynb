# Filters derivativos of first and segunda ordem ## About the code 1. Simula a signal of ECG with: * A onda senoidal of 1 Hz (representing the ritmo fundamental of the ECG). * A onda senoidal of 50 Hz (simulando noise of alta frequency). * A onda senoidal of 0,1 Hz (representing desvio of the line of base of baixa frequency). 2. Applies filtragem baseada in derivative: * The derivative of first ordem remove the desvio lento (efeito of filtragem passa-alta). * The derivative of segunda ordem aprimora the transies rpidas, facilitando the detection of the QRS. 3.plota the results: * ECG ruidoso original (with desvio of the line of base and noise). * First derivative (desvio of the line of base removido, form of onda more sharp). * Segunda derivative (aprimorando still more the recursos sharp).

```python
import numpy as np
import matplotlib.pyplot as plt

# Sampling frequencyfs = 500  # Hzt = np.linspace(0, 2, fs * 2, endpoint=False)  # 2 seconds of signal
# Simulated ECG-like waveform with noise and baseline driftecg_signal = np.sin(2 * np.pi * 1 * t) + 0.2 * np.sin(2 * np.pi * 50 * t)  # ECG + noisebaseline_drift = 0.5 * np.sin(2 * np.pi * 0.1 * t)  # Low-frequency driftecg_noisy = ecg_signal + baseline_drift  # Simulated raw ECG with drift
# Compute first-order and second-order derivativesecg_first_derivative = np.diff(ecg_noisy, n=1)
ecg_second_derivative = np.diff(ecg_noisy, n=2)

# Adjust time arrays for plotting (derivatives shorten the signal length)t_first = t[1:]
t_second = t[2:]

# Plot resultsplt.figure(figsize=(12, 6))

# Original Noisy ECG Signalplt.subplot(3, 1, 1)
plt.plot(t, ecg_noisy, label="Noisy ECG with Baseline Drift")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Simulated Noisy ECG Signal")
plt.legend()
plt.grid()

# First Derivative (High-Pass Effect)plt.subplot(3, 1, 2)
plt.plot(t_first, ecg_first_derivative, label="First-Order Derivative", color='r')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("First Derivative - Baseline Drift Removed")
plt.legend()
plt.grid()

# Second Derivative (Feature Enhancement)plt.subplot(3, 1, 3)
plt.plot(t_second, ecg_second_derivative, label="Second-Order Derivative", color='g')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Second Derivative - Enhancing QRS Features")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate, find_peaks

# --- Passo 1: Carregar o Sinal ECG ---ecg = np.loadtxt('ecg_hfn.dat')  # Carrega o sinal ECG do arquivofs = 1000                      # Taxa de amostragem em Hzt = np.arange(len(ecg)) / fs   # Cria o vetor de tempo
# Plota o ECG completo# # Plot the templateplt.plot(t*fs, ecg, label="Sinal de ECG original", color='red')
plt.title("Sinal de ECG original")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()

# --- Passo 2: Selecionar um Template de QRS ---# Escolha manualmente um segmento do ECG que contenha um complexo QRS representativo.# For example, assumimos that the ndices of 3120 the 3250 contm a QRS representativo.template_inicio = 3120         # ndice of incio (ajustar conforme necessary)template_fim = 3250           # ndice of fim (comprimento of the template = 100 samples)qrs_template = ecg[template_inicio:template_fim]

# # Plot the templateplt.plot(np.arange(qrs_template.shape[0]), qrs_template, label="Modelo de template de QRS (Full Cycle)", color='red')
plt.title("Modelo de complexo QRS")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()

# --- Passo 3: Detection of Batimentos via Correlation Cruzada ---# Calculates the correlation cruzada between the signal integer and the template of QRS.corr = correlate(ecg, qrs_template, mode='same')
corr = corr / np.max(corr)  # Normaliza the correlation
# Detecta picos in the correlation using a limiar.# The limiar and the distncia mnima between the picos (definida as 0,4 s) can to be ajustados.limiar = 0.4
picos, _ = find_peaks(corr, height=limiar, distance=int(limiar * fs))
print(f"ndices dos batimentos detectados: {picos}")

# --- Passo 4: Extrair Cycles Cardacos Completos in Torno of Each Batimento Detectado ---# Defines the window for cobrir a cycle cardaco completo.# For example, assumimos 300 ms before and 400 ms after the batimento detectado.amostras_pre = int(0.3 * fs)   # 300 ms antes do batimentoamostras_post = int(0.4 * fs)  # 400 ms after the batimentocomprimento_janela = amostras_pre + amostras_post  # Comprimento total da janela em amostras
# Extrai as janelas (apenas janelas que estejam completamente contidas no sinal)janelas = []
for pico in picos:
    inicio = pico - amostras_pre
    fim = pico + amostras_post
    if inicio >= 0 and fim <= len(ecg):
        janelas.append(ecg[inicio:fim])
janelas = np.array(janelas)
print(f"Foram extradas {janelas.shape[0]} janelas de comprimento {janelas.shape[1]} amostras.")

# --- Passo 5: To calculate the Mdia Sincronizada ---ciclo_medio = np.mean(janelas, axis=0)

# Cria um vetor de tempo para a janela, centrado no batimento detectadot_janela = np.linspace(-amostras_pre/fs, amostras_post/fs, comprimento_janela)

# --- Passo 6: Plotar os Resultados ---plt.figure(figsize=(12, 8))

# Plotar the complex QRS mdio (cobrindo a cycle cardaco completo)plt.subplot(2, 1, 1)
plt.plot(t_janela, ciclo_medio, label="Complexo QRS Mdio (Ciclo Completo)", color='red')
plt.title("Mdia Sincronizada dos Complexos QRS")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()

# Plotar a cycle of ECG amostral of the signal ruidoso for comparison# Here, plotamos the first window extrada as exampleplt.subplot(2, 1, 2)
plt.plot(t_janela, janelas[0], label="Ciclo de ECG Amostral", color='blue')
plt.title("Ciclo de ECG Amostral do Sinal Ruidoso")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
