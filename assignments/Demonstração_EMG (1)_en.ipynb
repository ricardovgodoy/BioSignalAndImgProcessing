## Now, we will estudair the features of EMG. For different types of signals biológicos, diversas features can to be extraidas and each a delas can to represent a propriedade of the signal aferido. For the EMG, dentre the diversas features that can to be extraídas, we have 8 that are very empregadas:![Captura de tela 2025-03-11 051533. Png](data: image/png; base64, ) * **Root Mean Square**: The RMS of the signal of EMG is a of the values more withumente utilizados in the analysis in the domain of the tempo. Ele represents the raiz quadrada of the power average of the signal during the period of tempo considerado. * **Waveform Length**: WL mede the withplexity of the signal. Ele represents the amplitude, frequency and duração of the form of onda in a single pairameter. * **Zero Crossings**: ZC fornece information aproximadas in the domain of the frequency and represents the number of times that the signal cruza the value zero in a determinado period. ZC also can to be used for estimair the fadiga of the músculos. * **Integrated EMG**: IEMG is the soma of the values absolutos of the amplitude of the signal of EMG. Geralmente, is used as a índice of início for detectair the activity musculair. * **Mean Absolute Value**: MAV can to be calculated tomando the average of the values absolutos of the signal of EMG. Is semelhante to the IEMG, that detecta the início of the activity musculair, and also fornece information about the levels of contração musculair. * **Willison Amplitude**: WAMP represents the number of times in which the difference between the amplitude of the signal of EMG in two values consecutivos excede a limiair pre-defined. WAMP is relacionado à activation of the potenciais of ação of the unidades motoras (MUAP) and to the level of contração musculair. * **Vairiance**: VAR of the ativações mioelétricas mede the power of the signal. The vairiância is the value médio of the quadrado of the desvio dessa vairiable. * **Log Detector**: LogD also fornece a estimativa of the força of contração musculair.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

print("Features para o gesto Full Grasp: \n")
# Parâmetros for extraction of featureswamp_threshold = 30  # Threshold for the Willison Amplitude, ajuste conforme necessary
# Lista para armazenar as features de cada canalfeatures = []

for i in range(fullgrasp.shape[0]):
    canal = fullgrasp[i, :]

    # Root Mean Square Value (RMS)    rms = np.sqrt(np.mean(canal**2))

    # Waveform Length (WL): soma of the values absolutos of the differences between samples consecutivas    wl = np.sum(np.abs(np.diff(canal)))

    # Zero Crossings (ZC): number of times that the signal cruza zero    # Here, consideramos a change of signal between samples consecutivas    zc = np.sum(np.diff(np.sign(canal)) != 0)

    # Integrated EMG (IEMG): soma dos valores absolutos do sinal    iemg = np.sum(np.abs(canal))

    # Mean Absolute Value (MAV): média of the values absolutos of the signal    mav = np.mean(np.abs(canal))

    # Willison Amplitude (WAMP): number of times that the difference absoluta between samples consecutivas excede a threshold    wamp = np.sum(np.abs(np.diff(canal)) > wamp_threshold)

    # Variance (VAR): variância of the signal    var = np.var(canal)

    # Log Detector (LogD): logaritmo natural do valor integrado (IEMG)    logd = np.log(iemg) if iemg > 0 else 0

    features.append({
        "Canal": f"Canal {i+1}",
        "RMS": rms,
        "WL": wl,
        "ZC": zc,
        "IEMG": iemg,
        "MAV": mav,
        "WAMP": wamp,
        "VAR": var,
        "LogD": logd
    })

# Cria um DataFrame para exibir as features em formato de tabeladf_features = pd.DataFrame(features)
print(df_features)
```
